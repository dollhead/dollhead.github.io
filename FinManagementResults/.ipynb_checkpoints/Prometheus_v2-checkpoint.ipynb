```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas import date_range,Series,DataFrame,read_csv, qcut
from pandas.tools.plotting import radviz,scatter_matrix,bootstrap_plot,parallel_coordinates
from numpy.random import randn
import numpy as np
#colorbrewer2 Dark2 qualitative color table
from pygeocoder import Geocoder
import wikipedia
import csv
import time
import shapefile
from sklearn.cluster import KMeans
from sklearn.cluster import MeanShift
import brewer2mpl
from matplotlib import rcParams
from pandasql import sqldf
import shapefile
pysqldf = lambda q: sqldf(q, globals())
#colorbrewer2 Dark2 qualitative color table
dark2_colors = brewer2mpl.get_map('Dark2', 'Qualitative', 7).mpl_colors

rcParams['figure.figsize'] = (10, 6)
rcParams['figure.dpi'] = 150
rcParams['axes.color_cycle'] = dark2_colors
rcParams['lines.linewidth'] = 2
rcParams['axes.facecolor'] = 'white'
rcParams['font.size'] = 14
rcParams['patch.edgecolor'] = 'white'
rcParams['patch.facecolor'] = dark2_colors[0]
rcParams['font.family'] = 'StixGeneral'

def remove_border(axes=None, top=False, right=False, left=True, bottom=True):
    """
    Minimize chartjunk by stripping out unnecesasry plot borders and axis ticks
    
    The top/right/left/bottom keywords toggle whether the corresponding plot border is drawn
    """
    ax = axes or plt.gca()
    ax.spines['top'].set_visible(top)
    ax.spines['right'].set_visible(right)
    ax.spines['left'].set_visible(left)
    ax.spines['bottom'].set_visible(bottom)
    
    #turn off all ticks
    ax.yaxis.set_ticks_position('none')
    ax.xaxis.set_ticks_position('none')
    
    #now re-enable visibles
    if top:
        ax.xaxis.tick_top()
    if bottom:
        ax.xaxis.tick_bottom()
    if left:
        ax.yaxis.tick_left()
    if right:
        ax.yaxis.tick_right()
```

```python
def geocode(addresses):
    for c in addresses:
        try:
            result = Geocoder.geocode(c)
            write_row_csv([c,result[0].raw[0]["formatted_address"],result[0].coordinates[0],result[0].coordinates[1],result[0].country,result[0].locality,result[0].administrative_area_level_1,result[0].administrative_area_level_2],'cities.csv')
            time.sleep(1)
        except Exception as e:
            write_row_csv([c,e],'erors.csv')
```

```python
def write_list_csv(data,filename,mode='a'):
    with open(filename, mode, newline='',encoding = 'UTF8') as csvfile:
        writer = csv.writer(csvfile, delimiter=',',
                                quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerows(data)
```

```python
def write_row_csv(data,filename):
    with open(filename, 'a', newline='',encoding = 'UTF8') as csvfile:
        writer = csv.writer(csvfile, delimiter=',',
                                quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerow(data)
```

```python
cities = []
errors = pd.read_csv('erors.csv')
for line in errors['﻿origin']:
    wikipedia.set_lang('uk')
    suggested = wikipedia.suggest(line)
    if suggested == None:
        wikipedia.set_lang('ru')
        if suggested == None:
            cities.append(line)
        else:
            cities.append(suggested)
    else:
        cities.append(suggested)
    write_row_csv([line,suggested],'suggested.csv')
```

```python
q = """
select distinct c.city,c.count,j.lat,j.lng,j.country,j.adm1
from
counts_cities c
join
joined j
on c.city = j.city
"""
city_counts_coord = pysqldf(q)
```

```python
cities_list = city_counts_coord.to_dict(outtype='records')
i = -1
while(i < len(cities_list) - 2):
    i += 1
    current = cities_list[i]['city']
    if cities_list[i+1]['city'] != current:
        continue
    else:
        while(cities_list[i+1]['city'] == current):
            cities_list.pop(i+1)
count_cities_df = pd.DataFrame(cities_list)
count_cities_df.to_csv("count_cities_df.csv",encoding = 'utf-8')
```

```python
q = """
select adm1,sum(count) suma
from uk_cities
group by adm1
order by suma desc
"""
regions_stats = pysqldf(q)
regions_stats
```
